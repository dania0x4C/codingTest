Î¨∏Ï†ú 5 (15Ï†ê) ‚Äì ÌñâÎ†¨Ïó∞ÏÇ∞ ÏÇ¨Ïö©

‚Ä¢ ÌñâÎ†¨ A, B Í∞Ä ÏïÑÎûòÏôÄ Í∞ôÏùÑ Îïå,

ÌñâÎ†¨ A, BÎäî Îã§ÏùåÍ≥º Í∞ôÎã§:

$$
A = \begin{bmatrix}
1 & -1 & 2 \\
2 & 1 & -3 \\
4 & 1 & 1
\end{bmatrix}, \quad
B = \begin{bmatrix}
3 & -1 & -2 \\
-4 & 2 & 1 \\
1 & 4 & -3
\end{bmatrix}
$$

‚Ä¢ (1) A+BÎ•º Íµ¨ÌïòÏó¨ ÌôîÎ©¥Ïóê ÌëúÏãúÌïòÍ∏∞

‚Ä¢ (2) (A x B)T ÌñâÎ†¨ÏùÑ Íµ¨ÌïòÏó¨ ÌôîÎ©¥Ïóê ÌëúÏãúÌïòÍ∏∞

‚Ä¢ (3) (A x B)Ïùò Ïó≠ÌñâÎ†¨ÏùÑ Íµ¨ÌïòÏó¨ ÌôîÎ©¥Ïóê ÌëúÏãúÌïòÍ∏∞

```python
import numpy as np

# ÌñâÎ†¨ A, B Ï†ïÏùò
A = np.array([
    [1, -1, 2],
    [2,  1, -3],
    [4,  1, 1]
])

B = np.array([
    [3,  -1, -2],
    [-4,  2, 1],
    [1,   4, -3]
])

# (1) A + B
sum_AB = A + B
print("(1) A + B:")
print(sum_AB)

# (2) (A x B)^T (Ï†ÑÏπò ÌñâÎ†¨)
product_AB = np.dot(A, B)  # ÎòêÎäî A @ B
transpose_product = product_AB.T
print("\n(2) (A x B)^T:")
print(transpose_product)

# (3) (A x B)^-1 (Ïó≠ÌñâÎ†¨)
try:
    inverse_product = np.linalg.inv(product_AB)
    print("\n(3) (A x B)^-1:")
    print(inverse_product)
except np.linalg.LinAlgError:
    print("\n(3) (A x B)^-1:")
    print("Ïó≠ÌñâÎ†¨Ïù¥ Ï°¥Ïû¨ÌïòÏßÄ ÏïäÏäµÎãàÎã§ (ÌñâÎ†¨Ïù¥ ÌäπÏù¥ÌñâÎ†¨ÏûÖÎãàÎã§).")
```
