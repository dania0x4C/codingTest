- 백트래킹(backtracking)은 문제를 해결하기 위해 여러 해결책을 시도하다가, 해결책이 실패할 경우 ```이전 단계로 되돌아가(백트래킹) 다른 해결책을 시도```하는 방법입니다. 이 방법은 잠재적인 해결책을 하나씩 조금씩 구축하면서, 해결책이 유효하거나 완전한지를 판단할 수 있게 되는 최종 단계에 도달할 때까지 계속 진행하는 시나리오에서 주로 사용됩니다.


- 백트래킹은 "시행착오"를 통해 문제를 해결하는 과정이며, 경우에 따라 매우 효율적일 수도 있고, 경우에 따라 많은 계산 시간을 요구할 수 있습니다. 이 방법은 ```종종 재귀 함수로 구현```되며, 각 단계에서 조건을 만족하지 못하면 자동으로 이전 단계로 되돌아가 다시 시도할 수 있습니다.
