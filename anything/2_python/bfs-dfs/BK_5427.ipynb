```python
class InputMock:
    def __init__(self, text: str):
        self.source = text.strip()

    def reset(self):
        self.lines = self.source.split('\n')
        self.current = 0

    def __call__(self):
        if not hasattr(self, 'lines') or self.current >= len(self.lines):
            self.reset()
        line = self.lines[self.current]
        self.current += 1
        return line

test_input = """
5
4 3
####
#*@.
####
7 6
###.###
#*#.#*#
#.....#
#.....#
#..@..#
#######
7 4
###.###
#....*#
#@....#
.######
5 5
.....
.***.
.*@*.
.***.
.....
3 3
###
#@#
###
"""

input = InputMock(test_input)
```

```python
from collections import deque

def solve_fire_escape(w, h, world):
    directions = [(0,1), (0,-1), (1,0), (-1,0)]
    fire_queue = deque()
    person_queue = deque()
    visited = [[False]*w for _ in range(h)]
    
    # 불, 사람 최초 좌표 찾기
    for i in range(h):
        for j in range(w):
            if world[i][j] == '@':
                person_queue.append((i, j, 0))
                visited[i][j] = True
            elif world[i][j] == '*':
                fire_queue.append((i, j))
                
    while person_queue:
        # 1️⃣ 불 확산
        for _ in range(len(fire_queue)):
            fx, fy = fire_queue.popleft()
            for dx, dy in directions:
                nx, ny = fx + dx, fy + dy
                if 0 <= nx < h and 0 <= ny < w and world[nx][ny] == '.':
                    world[nx][ny] = '*'  # 불 확산
                    fire_queue.append((nx, ny))

        # 2️⃣ 사람 이동
        for _ in range(len(person_queue)):
            x, y, cnt = person_queue.popleft()
            # 탈출 가능? (가장자리 도착 시 종료!)
            if x == 0 or x == h-1 or y == 0 or y == w-1:
                return cnt + 1
            for dx, dy in directions:
                nx, ny = x + dx, y + dy
                if 0 <= nx < h and 0 <= ny < w:
                    if world[nx][ny] == '.' and not visited[nx][ny]:
                        visited[nx][ny] = True
                        person_queue.append((nx, ny, cnt + 1))
    return "IMPOSSIBLE"

# === 실행 부분 ===
import sys
input = sys.stdin.readline
T = int(input().strip())
for _ in range(T):
    w, h = map(int, input().split())
    world = [list(input().strip()) for _ in range(h)]
    result = solve_fire_escape(w, h, world)
    print(result)
```
