```python
import sys
N, M, x, y, k = map(int, sys.stdin.readline().strip().split())

world = [list(map(int, sys.stdin.readline().strip().split())) for _ in range(N)]
directions =list(map(int, sys.stdin.readline().strip().split()))
dice = [0] * 7

def turnDice(dice, direction):
    temp = dice[:]
    if direction == 1:
        dice[1] = temp[4]
        dice[3] = temp[1]
        dice[6] = temp[3]
        dice[4] = temp[6]
    elif direction == 2:
        dice[1] = temp[3]
        dice[4] = temp[1]
        dice[6] = temp[4]
        dice[3] = temp[6]
    elif direction == 3:
        dice[1] = temp[5]
        dice[2] = temp[1]
        dice[6] = temp[2]
        dice[5] = temp[6]
    elif direction == 4:
        dice[1] = temp[2]
        dice[5] = temp[1]
        dice[6] = temp[5]
        dice[2] = temp[6]

def checkDice(x, y, world, dice):
    if world[x][y] == 0:
        world[x][y] = dice[6]
    else:
        dice[6] = world[x][y]
        world[x][y] = 0

def move(world, dice, direction, x, y):
    if direction == 1:
        dx, dy = x, y + 1
        if dy < M:
            turnDice(dice, direction)
            checkDice(dx, dy, world, dice)
            return dx, dy
    elif direction == 2:
        dx, dy = x, y - 1
        if 0 <= dy:
            turnDice(dice, direction)
            checkDice(dx, dy, world, dice)
            return dx, dy
    elif direction == 3:
        dx, dy = x - 1, y
        if 0 <= dx:
            turnDice(dice, direction)
            checkDice(dx, dy, world, dice)
            return dx, dy
    else:
        dx, dy = x + 1, y
        if dx < N:
            turnDice(dice, direction)
            checkDice(dx, dy, world, dice)
            return dx, dy

    return x, y
for direction in directions:
    nx, ny = move(world, dice, direction, x, y)
    if(nx, ny) != (x, y):
        print(dice[1])
        x, y = nx, ny
```
