```python
import sys
from collections import deque

N = int(sys.stdin.readline().strip())
minusNumbers = deque()
plusNumbers = deque()
zero = False
result = 0
plus = 0
minus = 0

for i in range(N):
    a = int(sys.stdin.readline().strip())
    if a > 0:
        plusNumbers.append(a)
    elif a < 0:
        minusNumbers.append(a)
    else:
        zero = True

sortedP = sorted(plusNumbers, reverse = True)
sortedM = sorted(minusNumbers)

for i in range(0, len(sortedP) - 1, 2):
    result += sortedP[i] * sortedP[i+1]
if(len(sortedP) % 2 == 1):
    plus = sortedP[-1]

for i in range(0, len(sortedM) - 1, 2):
    result += sortedM[i] * sortedM[i+1]
if(len(sortedM) % 2 == 1):
    minus = sortedM[-1]

if zero:
    if minus:
        result+=plus
else:
    if plus:
        if minus:
            result += plus * minus
        else:
            result += plus
    else:
        result += minus

print(result)
```

```python
import sys

N = int(sys.stdin.readline().strip())
plus_numbers = []
minus_numbers = []
zero = False
ones_count = 0
result = 0

for _ in range(N):
    num = int(sys.stdin.readline().strip())
    if num > 1:
        plus_numbers.append(num)
    elif num == 1:
        ones_count += 1
    elif num == 0:
        zero = True
    else:
        minus_numbers.append(num)

plus_numbers.sort(reverse=True)

minus_numbers.sort()

i = 0
while i < len(plus_numbers) - 1:
    result += plus_numbers[i] * plus_numbers[i+1]
    i += 2

if i < len(plus_numbers):
    result += plus_numbers[i]

i = 0
while i < len(minus_numbers) - 1:
    result += minus_numbers[i] * minus_numbers[i+1]
    i += 2

if i < len(minus_numbers):
    if not zero:
        result += minus_numbers[i]

result += ones_count

print(result)
```
