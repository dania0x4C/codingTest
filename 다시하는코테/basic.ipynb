# BFS

```python
# bfs 1ì°¨ì›
from collections import deque

def bfs_one_dimension(N, K):
    MAX = 100001
    dist = [-1] * MAX  
    
    q = deque()
    
    q.append(N)
    dist[N] = 0
    
    while q:
        v = q.popleft()

        if v == K:
            return dist[v]
        
        next_steps = [v + 1, v - 1, v * 2] 

        for next_v in next_steps:
            
            if 0 <= next_v < MAX:
                if dist[next_v] == -1:
                    
                    dist[next_v] = dist[v] + 1
                    q.append(next_v)
    
    return -1
```

```python
# bfs ì¥ì• ë¬¼
from collections import deque

# ì „ì—­ ë³€ìˆ˜ ì„¤ì • (ì˜ˆì‹œ)
R, C = 5, 5
# graph: 1ì€ í†µë¡œ, 0ì€ ë²½ì´ë¼ê³  ê°€ì •
graph = [[int(x) for x in '11011'], [int(x) for x in '10101'], [int(x) for x in '11111'], [int(x) for x in '00001'], [int(x) for x in '11111']]
direction = ((1,0), (-1, 0), (0, 1), (0, -1))

# visited[r][c][k]: (r, c) ì¢Œí‘œì— kë²ˆì˜ ë²½ì„ ë¶€ìˆ˜ê³  ë„ì°©í–ˆëŠ”ì§€ ê¸°ë¡
# k=0: ë²½ì„ 0ë²ˆ ë¶€ìˆ¨, k=1: ë²½ì„ 1ë²ˆ ë¶€ìˆ¨
visited = [[[0] * 2 for _ in range(C)] for _ in range(R)] 

def bfs_state_space(start_r, start_c, max_break=1):
    # í: (r, c, í˜„ì¬ê¹Œì§€ì˜ ê±°ë¦¬, ë²½ì„ ë¶€ìˆœ íšŸìˆ˜)
    q = deque([(start_r, start_c, 1, 0)]) 
    visited[start_r][start_c][0] = 1 # ì‹œì‘ ìœ„ì¹˜ ë°©ë¬¸ ì²˜ë¦¬ (ê±°ë¦¬ 1)
    
    while q:
        r, c, dist, broken = q.popleft()
        
        # ëª©í‘œ ì§€ì  ë„ì°© ì¡°ê±´ (ì˜ˆì‹œ)
        # if r == R-1 and c == C-1: return dist

        for dr, dc in direction:
            nr, nc = r + dr, c + dc
            
            if 0 <= nr < R and 0 <= nc < C:
                
                # Case 1: ë‹¤ìŒ ì¹¸ì´ í†µë¡œì¼ ë•Œ (graph[nr][nc] == 1)
                if graph[nr][nc] == 1:
                    # 3ì°¨ì› visited: í˜„ì¬ ë¶€ìˆœ íšŸìˆ˜ ìƒíƒœ(broken)ë¡œ ì•„ì§ ë°©ë¬¸í•˜ì§€ ì•Šì€ ê²½ìš°
                    if visited[nr][nc][broken] == 0:
                        visited[nr][nc][broken] = dist + 1
                        q.append((nr, nc, dist + 1, broken))
                
                # Case 2: ë‹¤ìŒ ì¹¸ì´ ë²½ì¼ ë•Œ (graph[nr][nc] == 0)
                elif graph[nr][nc] == 0:
                    # ë²½ì„ ë¶€ìˆ  ê¸°íšŒê°€ ë‚¨ì•„ ìˆê³  (broken < max_break)
                    # í•´ë‹¹ ìƒíƒœ(broken+1)ë¡œ ì•„ì§ ë°©ë¬¸í•˜ì§€ ì•Šì€ ê²½ìš°
                    if broken < max_break and visited[nr][nc][broken + 1] == 0:
                        visited[nr][nc][broken + 1] = dist + 1
                        q.append((nr, nc, dist + 1, broken + 1))
                        
    # ìµœë‹¨ ê±°ë¦¬ë¥¼ ì°¾ì§€ ëª»í•˜ë©´ -1 ë°˜í™˜
    return -1
```

```python
# DFS ê¸°ë³¸
import sys
sys.setrecursionlimit(2000) 

R, C = 5, 5
graph = [[1]*C for _ in range(R)] # 1ì€ í†µë¡œ, 0ì€ ë²½
visited = [[False] * C for _ in range(R)]
direction = ((1,0), (-1, 0), (0, 1), (0, -1))

def dfs(r, c):
    # 1. ê²½ê³„ ì¡°ê±´ ë° ì¢…ë£Œ ì¡°ê±´
    if not (0 <= r < R and 0 <= c < C):
        return
    if graph[r][c] == 0: # ë²½(ì¥ì• ë¬¼)ì¼ ê²½ìš° ì¢…ë£Œ
        return
    if visited[r][c]:
        return
    
    # 2. í˜„ì¬ ë…¸ë“œ ì²˜ë¦¬ ë° ë°©ë¬¸ ì²˜ë¦¬
    visited[r][c] = True
    # (ì—¬ê¸°ì— í•„ìš”í•œ ë¡œì§ ì¶”ê°€: count ì¦ê°€, ê°’ ê¸°ë¡ ë“±)
    
    # 3. ì´ì›ƒ ë…¸ë“œ ì¬ê·€ í˜¸ì¶œ (ê¹Šì´ ìš°ì„ )
    for dr, dc in direction:
        nr, nc = r + dr, c + dc
        dfs(nr, nc)
        
    # (ì„ íƒ: ë°±íŠ¸ë˜í‚¹ ë¡œì§ ì¶”ê°€)
    # visited[r][c] = False
```

```python
# DFS ì˜ì—­ í¬ê¸°
def dfs_area_size(r, c, graph, visited):
    R, C = len(graph), len(graph[0])
    
    # 1. ê²½ê³„/ë°©ë¬¸/ì¡°ê±´ í™•ì¸
    if not (0 <= r < R and 0 <= c < C): return 0
    if graph[r][c] == 0 or visited[r][c]: return 0 # 0ì´ ë²½ì´ë¼ ê°€ì •
    
    visited[r][c] = True
    size = 1 # í˜„ì¬ ë…¸ë“œ í¬í•¨
    
    # 2. ëª¨ë“  ì¸ì ‘ ë…¸ë“œì— ëŒ€í•´ ì¬ê·€ í˜¸ì¶œ
    for dr, dc in direction:
        size += dfs_area_size(r + dr, c + dc, graph, visited)
        
    return size
```

```python
# DFS ë°±íŠ¸ë™í‚¹
def dfs_backtracking(level, current_sum, target, numbers, visited):
    # ì¢…ë£Œ ì¡°ê±´: ëª¨ë“  ìˆ«ìë¥¼ ì‚¬ìš©í–ˆê±°ë‚˜(level == N), ì›í•˜ëŠ” í•©(target)ì— ë„ë‹¬í–ˆì„ ë•Œ
    if level == len(numbers):
        if current_sum == target:
            return 1 # ê²½ìš°ì˜ ìˆ˜ 1ê°œ ë°œê²¬
        return 0

    result = 0
    
    # 1. í˜„ì¬ ìˆ«ìë¥¼ í¬í•¨í•˜ëŠ” ê²½ìš°
    result += dfs_backtracking(level + 1, current_sum + numbers[level], target, numbers, visited)
    
    # 2. í˜„ì¬ ìˆ«ìë¥¼ í¬í•¨í•˜ì§€ ì•ŠëŠ” ê²½ìš° (ë˜ëŠ” ë‹¤ë¥¸ ì„ íƒì§€)
    result += dfs_backtracking(level + 1, current_sum - numbers[level], target, numbers, visited)
    
    return result
```

# DP - ì˜ ëª°ë¼ì„œ ë¬¸ì œ í’€ì–´ì•¼í• ë“¯

```python
# bottom - up 1ì°¨ì›
# ë¬¸ì œ: Në²ˆì§¸ ê³„ë‹¨ì„ ì˜¤ë¥´ëŠ” ë°©ë²•ì˜ ìˆ˜
def dp_1d_basic(N):
    # D[i]: ië²ˆì§¸ ê³„ë‹¨ì„ ì˜¤ë¥´ëŠ” ë°©ë²•ì˜ ìˆ˜
    D = [0] * (N + 1)
    
    # ê¸°ì € ì¡°ê±´ (Base Cases)
    if N >= 1:
        D[1] = 1
    if N >= 2:
        D[2] = 2 # 1+1, 2ì¹¸ ì í”„
        
    # ì í™”ì‹ (í˜„ì¬ ìƒíƒœëŠ” ì´ì „ ìƒíƒœì˜ í•©)
    for i in range(3, N + 1):
        # D[i] = D[i-1] + D[i-2] (1ì¹¸ ì „ì—ì„œ ì˜¤ê±°ë‚˜ 2ì¹¸ ì „ì—ì„œ ì˜¤ê±°ë‚˜)
        D[i] = D[i-1] + D[i-2] 
        
    return D[N]
```

```python
# ë¬¸ì œ: Nê°œì˜ ë¬¼ê±´ ì¤‘ ìš©ëŸ‰(W)ì„ ë„˜ì§€ ì•Šê²Œ ìµœëŒ€ ê°€ì¹˜(V)ë¥¼ êµ¬í•˜ë¼ (ëƒ…ìƒ‰)
def dp_2d_knapsack(N, max_W, items):
    # D[i][j]: ië²ˆì§¸ ë¬¼ê±´ê¹Œì§€ ê³ ë ¤í–ˆì„ ë•Œ, ë¬´ê²Œ jì—ì„œì˜ ìµœëŒ€ ê°€ì¹˜
    D = [[0] * (max_W + 1) for _ in range(N + 1)]
    
    for i in range(1, N + 1):
        weight, value = items[i-1]
        
        for j in range(1, max_W + 1):
            
            # 1. í˜„ì¬ ë¬¼ê±´ì„ ë„£ì„ ìˆ˜ ì—†ëŠ” ê²½ìš° (ë¬´ê²Œê°€ jë³´ë‹¤ í¼)
            if weight > j:
                D[i][j] = D[i-1][j] # ì´ì „ ë¬¼ê±´ê¹Œì§€ì˜ ê°€ì¹˜ì™€ ë™ì¼
            
            # 2. í˜„ì¬ ë¬¼ê±´ì„ ë„£ì„ ìˆ˜ ìˆëŠ” ê²½ìš°
            else:
                # D[i][j] = max(â‘ ì•ˆ ë„£ì—ˆì„ ë•Œ, â‘¡ë„£ì—ˆì„ ë•Œ)
                D[i][j] = max(D[i-1][j], D[i-1][j - weight] + value)
                
    return D[N][max_W]
```

```python
import sys
# top - down 
sys.setrecursionlimit(2000) 

# ë©”ëª¨ ë°°ì—´ (DP í…Œì´ë¸”)ì„ ì „ì—­ ë˜ëŠ” í´ë˜ìŠ¤ ë©¤ë²„ ë³€ìˆ˜ë¡œ ì´ˆê¸°í™”
# 0ì€ ì•„ì§ ê³„ì‚°ë˜ì§€ ì•ŠìŒì„ ì˜ë¯¸í•©ë‹ˆë‹¤.
memo = [0] * 101 

def dp_top_down(n):
    # 1. ğŸš¨ ê¸°ì € ì¡°ê±´ (Base Case)
    if n == 1 or n == 2:
        return 1
    
    # 2. ğŸ”‘ ë©”ëª¨ì´ì œì´ì…˜ (Memoization): ì´ë¯¸ ê³„ì‚°ëœ ê°’ì´ë©´ ì¦‰ì‹œ ë°˜í™˜
    if memo[n] != 0:
        return memo[n]
        
    # 3. ì¬ê·€ í˜¸ì¶œ ë° ë©”ëª¨ ì €ì¥
    # ëª©í‘œë¥¼ ë‹¬ì„±í•˜ê¸° ìœ„í•´ í•˜ìœ„ ë¬¸ì œ(n-1, n-2)ë¥¼ í˜¸ì¶œ
    memo[n] = dp_top_down(n - 1) + dp_top_down(n - 2)
    
    return memo[n]

# ì˜ˆì‹œ ì‹¤í–‰
# result = dp_top_down(10)
# print(f"D[10] ê°’: {result}")
```

```python
# ë¬¸ì œ: Nì„ 1ë¡œ ë§Œë“œëŠ” ìµœì†Œ íšŸìˆ˜ (1ì°¨ì› ë°°ì—´ ìœ í˜•)
# memo ë°°ì—´ì€ í•„ìš”í•œ ë§Œí¼ ë¯¸ë¦¬ ì´ˆê¸°í™”ë˜ì—ˆë‹¤ê³  ê°€ì •í•©ë‹ˆë‹¤.

def dp_min_cost_top_down(n):
    # 1. ê¸°ì € ì¡°ê±´
    if n == 1:
        return 0 # 1ì€ 0ë²ˆ ë§Œì— 1ì´ ë¨
    
    # 2. ë©”ëª¨ì´ì œì´ì…˜
    if memo[n] != 0:
        return memo[n]
        
    # 3. ìµœì†Œê°’ ê³„ì‚°: (n-1)ì—ì„œ ì˜¤ëŠ” ê²½ìš°ë¥¼ ê¸°ë³¸ê°’ìœ¼ë¡œ ì„¤ì •
    min_ops = dp_min_cost_top_down(n - 1) + 1
    
    # 4. 2ë¡œ ë‚˜ëˆ„ëŠ” ê²½ìš° í™•ì¸
    if n % 2 == 0:
        min_ops = min(min_ops, dp_min_cost_top_down(n // 2) + 1)
        
    # 5. 3ìœ¼ë¡œ ë‚˜ëˆ„ëŠ” ê²½ìš° í™•ì¸
    if n % 3 == 0:
        min_ops = min(min_ops, dp_min_cost_top_down(n // 3) + 1)
        
    # 6. ë©”ëª¨ ì €ì¥ ë° ë°˜í™˜
    memo[n] = min_ops
    return min_ops
```

```python
# ë¬¸ì œ: ì£¼ì–´ì§„ í™œë™ ì¤‘ ìµœëŒ€í•œ ë§ì€ í™œë™ì„ ì„ íƒí•˜ë¼
# activities: [[ì‹œì‘ì‹œê°„, ì¢…ë£Œì‹œê°„], ...]
def greedy_activity_selection(activities):
    
    # 1. ğŸš¨ í•µì‹¬: ì •ë ¬ ê¸°ì¤€ ì°¾ê¸° ğŸš¨
    # ë¹¨ë¦¬ ëë‚˜ëŠ” í™œë™ì„ ë¨¼ì € ì„ íƒí•´ì•¼ ë‹¤ìŒ í™œë™ì„ í•  ê¸°íšŒê°€ ë§ì•„ì§‘ë‹ˆë‹¤.
    activities.sort(key=lambda x: x[1]) # ì¢…ë£Œ ì‹œê°„ì„ ê¸°ì¤€ìœ¼ë¡œ ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬

    count = 0
    last_end_time = 0
    
    for start, end in activities:
        # 2. í˜„ì¬ í™œë™ì´ ì´ì „ í™œë™ì— ë°©í•´ë°›ì§€ ì•ŠëŠ”ë‹¤ë©´ ì„ íƒ
        if start >= last_end_time:
            count += 1
            last_end_time = end # í™œë™ì„ ì„ íƒí–ˆìœ¼ë¯€ë¡œ ì¢…ë£Œ ì‹œê°„ì„ ê°±ì‹ 
            
    return count
```

```python
# ë¬¸ì œ: ì£¼ì–´ì§„ ê¸ˆì•¡ì„ ìµœì†Œ í™”í ë‹¨ìœ„ë¡œ ê±°ìŠ¬ëŸ¬ ì£¼ë¼
def greedy_coin_change(amount, coins):
    
    # 1. í™”í ë‹¨ìœ„ë¥¼ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬ (ê°€ì¥ í° ë‹¨ìœ„ë¶€í„° ì²˜ë¦¬)
    coins.sort(reverse=True)
    
    count = 0
    
    for coin in coins:
        # 2. í˜„ì¬ í™”í ë‹¨ìœ„ë¡œ ëª‡ ê°œë¥¼ ì‚¬ìš©í•  ìˆ˜ ìˆëŠ”ì§€ ê³„ì‚°
        num_coins = amount // coin
        count += num_coins
        
        # 3. ë‚¨ì€ ê¸ˆì•¡ ê°±ì‹ 
        amount %= coin
        
        if amount == 0:
            break
            
    return count
```

# Dijkstra

```python
import heapq
import sys

# ë¬´í•œëŒ€ ê°’ ì„¤ì • (ì¶©ë¶„íˆ í° ê°’)
INF = int(1e9) 

def dijkstra(start_node, N, graph):
    # 1. ê±°ë¦¬ ë°°ì—´ ì´ˆê¸°í™”
    dist = [INF] * (N + 1)
    q = []
    
    # 2. ì‹œì‘ ë…¸ë“œ ì´ˆê¸°í™”: (ë¹„ìš©, ë…¸ë“œë²ˆí˜¸) í˜•íƒœë¡œ íì— ì‚½ì…
    heapq.heappush(q, (0, start_node))
    dist[start_node] = 0
    
    while q:
        # 3. í˜„ì¬ íì—ì„œ ê°€ì¥ ë¹„ìš©ì´ ì ì€ ë…¸ë“œ ì¶”ì¶œ
        cost, now = heapq.heappop(q)
        
        # ì´ë¯¸ ì²˜ë¦¬ëœ ë…¸ë“œ(í˜„ì¬ ê¸°ë¡ëœ distë³´ë‹¤ ë¹„ìš©ì´ í° ê²½ìš°)ëŠ” ë¬´ì‹œ
        if dist[now] < cost:
            continue
            
        # 4. í˜„ì¬ ë…¸ë“œì™€ ì—°ê²°ëœ ì¸ì ‘ ë…¸ë“œ íƒìƒ‰
        for next_node, weight in graph[now]:
            
            new_cost = cost + weight
            
            # 5. ìµœë‹¨ ê±°ë¦¬ ê°±ì‹  ë¡œì§
            if new_cost < dist[next_node]:
                dist[next_node] = new_cost
                # ìµœë‹¨ ê±°ë¦¬ê°€ ê°±ì‹ ëœ ë…¸ë“œë§Œ íì— ë‹¤ì‹œ ì‚½ì…
                heapq.heappush(q, (new_cost, next_node))
                
    return dist
    
```

# Kruskal

```python
# 1. Union-Findë¥¼ ìœ„í•œ ë¶€ëª¨ í…Œì´ë¸” ì´ˆê¸°í™”
# parent[i] = i: ì´ˆê¸°ì—ëŠ” ëª¨ë“  ë…¸ë“œê°€ ìê¸° ìì‹ ì„ ëŒ€í‘œ ë…¸ë“œë¡œ ê°€ì§
parent = []

def find_parent(x):
    # ë£¨íŠ¸ ë…¸ë“œê°€ ì•„ë‹ˆë¼ë©´, ë£¨íŠ¸ ë…¸ë“œë¥¼ ì°¾ì„ ë•Œê¹Œì§€ ì¬ê·€ í˜¸ì¶œ (ê²½ë¡œ ì••ì¶• ì ìš©)
    if parent[x] == x:
        return x
    parent[x] = find_parent(parent[x]) # â­ ê²½ë¡œ ì••ì¶•: ì°¾ì€ ë£¨íŠ¸ë¥¼ ë°”ë¡œ ë¶€ëª¨ë¡œ ì„¤ì •
    return parent[x]

def union_parent(a, b):
    # ê° ì›ì†Œê°€ ì†í•œ ì§‘í•©ì˜ ë£¨íŠ¸ë¥¼ ì°¾ìŒ
    root_a = find_parent(a)
    root_b = find_parent(b)
    
    # ë‘ ë£¨íŠ¸ê°€ ë‹¤ë¥´ë©´ (ë‹¤ë¥¸ ì§‘í•©ì´ë¼ë©´) í•©ì¹¨
    if root_a != root_b:
        # ì¼ë°˜ì ìœ¼ë¡œ ì‘ì€ ë²ˆí˜¸ì˜ ë£¨íŠ¸ë¥¼ ë¶€ëª¨ë¡œ ì„¤ì •
        if root_a < root_b:
            parent[root_b] = root_a
        else:
            parent[root_a] = root_b
        return True # í•©ì¹˜ê¸° ì„±ê³µ
    return False # ì´ë¯¸ ê°™ì€ ì§‘í•© (ì‚¬ì´í´ ë°œìƒ)
```

```python
def kruskal_algorithm(N, edges):
    """
    N: ë…¸ë“œì˜ ê°œìˆ˜, edges: ì „ì²´ ê°„ì„  ë¦¬ìŠ¤íŠ¸ [(ë¹„ìš©, ë…¸ë“œA, ë…¸ë“œB), ...]
    """
    global parent
    
    # 1. Union-Find ì´ˆê¸°í™”
    parent = [i for i in range(N + 1)] # ë…¸ë“œ ë²ˆí˜¸ 1ë¶€í„° Nê¹Œì§€ ì‚¬ìš© ê°€ì •
    
    # 2. ê°„ì„ ì„ ë¹„ìš© ìˆœìœ¼ë¡œ ì •ë ¬ (ê°€ì¥ ë‚®ì€ ë¹„ìš©ë¶€í„°)
    edges.sort()
    
    total_cost = 0
    num_edges = 0
    
    for cost, a, b in edges:
        
        # 3. Union-Findë¥¼ ì‚¬ìš©í•˜ì—¬ ì‚¬ì´í´ ë°œìƒ ì—¬ë¶€ í™•ì¸
        # (ë‘ ë…¸ë“œ a, bì˜ ë£¨íŠ¸ê°€ ë‹¤ë¥´ë©´, ì•„ì§ ì—°ê²°ë˜ì§€ ì•Šì€ ë‹¤ë¥¸ ì§‘í•©ì„)
        if find_parent(a) != find_parent(b):
            
            # 4. ì‚¬ì´í´ì´ ì—†ìœ¼ë©´ Union ìˆ˜í–‰ ë° ë¹„ìš© í•©ì‚°
            union_parent(a, b)
            total_cost += cost
            num_edges += 1
            
            # 5. MST ì™„ì„± ì¡°ê±´: ê°„ì„  ìˆ˜ê°€ (ë…¸ë“œ ìˆ˜ - 1)ì´ ë˜ë©´ ì¢…ë£Œ
            if num_edges == N - 1:
                break
                
    return total_cost
```

```python
# íŠ¹ì • ë°°ì—´ì—ì„œ ì—°ì†ë˜ëŠ” ê°’ìœ¼ë¡œ íŠ¹ì • ê°’ì´ ë˜ëŠ”ì§€ë¥¼ í™•ì¸í•˜ëŠ” ë°©ë²•ì´ë‹¤.
def two_pointers_subarray_sum(data, M):
    N = len(data)
    # í˜„ì¬ ë¶€ë¶„ ë°°ì—´ì˜ ì‹œì‘ ì¸ë±ìŠ¤
    start = 0
    # í˜„ì¬ ë¶€ë¶„ ë°°ì—´ì˜ ë ì¸ë±ìŠ¤ ë‹¤ìŒ ìœ„ì¹˜
    end = 0
    # í˜„ì¬ ë¶€ë¶„ ë°°ì—´ì˜ í•©
    current_sum = 0
    # í•©ì´ Mì´ ë˜ëŠ” ë¶€ë¶„ ë°°ì—´ì˜ ê°œìˆ˜
    count = 0
    while True:
        # 1. í•©ì´ ëª©í‘œê°’ Më³´ë‹¤ í¬ê±°ë‚˜ ê°™ìœ¼ë©´, start(ì™¼ìª½)ë¥¼ ì´ë™ì‹œì¼œ í•©ì„ ì¤„ì¸ë‹¤.
        if current_sum >= M:
            current_sum -= data[start]
            start += 1
        # 2. end(ì˜¤ë¥¸ìª½)ê°€ ë°°ì—´ ëì— ë„ë‹¬í–ˆìœ¼ë©´ ì¢…ë£Œí•œë‹¤.
        elif end == N:
            break
        # 3. í•©ì´ ëª©í‘œê°’ Më³´ë‹¤ ì‘ìœ¼ë©´, end(ì˜¤ë¥¸ìª½)ë¥¼ ì´ë™ì‹œì¼œ í•©ì„ ëŠ˜ë¦°ë‹¤.
        else:
            current_sum += data[end]
            end += 1

        # 4. í•©ì´ Mê³¼ ì •í™•íˆ ì¼ì¹˜í•˜ëŠ” ê²½ìš° ì¹´ìš´íŠ¸
        if current_sum == M:
            count += 1

    return count

# ì˜ˆì‹œ ì‹¤í–‰: arr = [1, 2, 3, 4, 2, 5], target = 5
# result = two_pointers_subarray_sum(arr, 5) # ê²°ê³¼: 3
```

```python
# Prefix Sum: 1ì°¨ì› ë°°ì—´ì˜ êµ¬ê°„ í•© (O(1) ì¿¼ë¦¬)
def prefix_sum_1d(arr):
    N = len(arr)
    # ëˆ„ì  í•© ë°°ì—´ P ìƒì„±. P[0] = 0ìœ¼ë¡œ ì´ˆê¸°í™”í•˜ì—¬ 1-based indexì²˜ëŸ¼ ì‚¬ìš© ì¤€ë¹„
    # P[i] = arr[0] + ... + arr[i-1]
    P = [0] * (N + 1)

    # 1. ëˆ„ì  í•© ë°°ì—´ ìƒì„± (O(N))
    for i in range(N):
        P[i+1] = P[i] + arr[i]

    return P

def get_range_sum(P, i, j):
    """
    P: ëˆ„ì  í•© ë°°ì—´
    i, j: í•©ì„ êµ¬í•˜ê³  ì‹¶ì€ ë°°ì—´ arrì˜ 0-based ì¸ë±ìŠ¤ (arr[i]ë¶€í„° arr[j]ê¹Œì§€)
    """
    # 2. êµ¬ê°„ í•© ê³„ì‚° (O(1))
    # arr[i]ë¶€í„° arr[j]ê¹Œì§€ì˜ í•©ì€ P[j+1] - P[i]
    return P[j + 1] - P[i]

# ì˜ˆì‹œ ì‹¤í–‰: data = [10, 20, 30, 40, 50]
# P = prefix_sum_1d(data) # P: [0, 10, 30, 60, 100, 150]
# sum_range = get_range_sum(P, 1, 3) # data[1] + data[2] + data[3] = 90
```
