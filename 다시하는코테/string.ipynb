```python
class InputMock:
    def __init__(self, text: str):
        self.source = text.strip()

    def reset(self):
        self.lines = self.source.split('\n')
        self.current = 0

    def __call__(self):
        if not hasattr(self, 'lines') or self.current >= len(self.lines):
            self.reset()
        line = self.lines[self.current]
        self.current += 1
        return line

test_input = """
abababababababa
"""

input = InputMock(test_input)
```

1. 원형의 문자열 상태를 확인하는 방법으로는 문자열을 한번 더 붙이는 방법이 존재 ex) ss = s + s
2. a가 전부 붙어있는 경우의 최대 개수를 확인해서 window 구간으로 사용
3. 그 안에서 b가 제일 적은수가 있을때 교환의 최솟값이다.

```python
import sys

s = input()
n = len(s)
# 다음의 함수로 문자열 내부의 특정 값 개수 확인 가능
window_size = s.count('a')
ss = s + s


# 예외 처리: a가 없으면 교환할 필요 없음
if window_size == 0:
    print(0)
    sys.exit()

# 초기 윈도우 사이즈 까지의 b의 값을 확인
current_b = ss[:window_size].count('b')
min_b = current_b

# 윈도우의 앞과 끝을 확인해서 b의 입출력을 감지 후 반영
for i in range(n):

    if ss[i] == 'b':
        current_b -= 1

    if ss[i + window_size] == 'b':
        current_b += 1
    
    min_b = min(min_b, current_b)

print(min_b)
```
