```python
# 백준-style input()을 테스트용으로 흉내내기 위한 클래스
class InputMock:
    def __init__(self, text: str):
        # 사용자가 지정한 입력 문자열을 저장 (전체 문자열)
        self.source = text.strip()  # 앞뒤 공백 제거
        # 아직 줄 단위로 나누지는 않음 → 첫 호출 시 reset()으로 분리됨

    def reset(self):
        # 원본 문자열을 줄 단위 리스트로 변환
        self.lines = self.source.split('\n')
        # 현재 읽고 있는 줄 번호를 초기화
        self.current = 0

    def __call__(self):
        # 아직 reset된 적 없거나, 모든 줄을 다 읽었으면 자동 초기화
        if not hasattr(self, 'lines') or self.current >= len(self.lines):
            self.reset()
        # 현재 줄 반환 후, 다음 줄로 인덱스 이동
        line = self.lines[self.current]
        self.current += 1
        return line


# 테스트 입력 정의 (여러 줄 입력도 가능)
test_input = """
4 8
3 1 2 3 4 1 1 2
"""

# input()처럼 사용 가능하게 인스턴스 생성
input = InputMock(test_input)
```

```python
import sys

H, W = map(int, input().strip().split())
water = list(map(int, input().strip().split()))
total = 0

for i in range(1, W-1):
    left = max(water[:i])
    right = max(water[i+1:])
    hight = min(left, right)
    if hight > water[i]:
        total += hight - water[i]

print(total)
```

1. 내가 서있는 곳을 기준으로 옆으로 한칸식 이동
2. 서있는 기준으로 왼쪽과 오른쪽에서 가장 큰 벽을 확인 => 그중 낮은 벽까지 내가 있는 위치에서 물이 차게 됨
3. 내가 서 있는 위치의 벽의 높이와 가장 큰 공간의 높이차를 빼면 물이 찰 높이의 크기를 확인 가능
