```python
import sys
from collections import deque

input = sys.stdin.readline
N = int(input().strip())
K = int(input().strip())

board = [[0] * N for i in range(N)]
board[0][0] = 1

for _ in range(K):
    apple_x, apple_y = map(int, sys.stdin.readline().strip().split())
    board[apple_x][apple_y] = 2

L = int(input().strip())
location = deque()
location.append([0, 0])

time = 0
direction = 'R'
C = 'D'
x, y = 0, 0

# 1 = 뱀
# 2 = 사과
# 0 = 빈 곳
# 이동한 board의 상태에 따라 뱀의 상태와 board의 상태를 update
def checkBoard(board, x, y):
    if 0 <= x < N and 0 <= y < N:

        if board[x][y] == 0:
            sx, sy = location.pop()
            board[sx, sy] = 0
            location.append([x, y])
            board[x, y] = 1
            return 1
        elif board[x][y] == 2:
            location.append([x, y])
            board[x, y] = 1
            return 1
        else:
            return 0
    else:
        return 0

# 뱀을 이동 시키고 게임 탈락시 time의 print
def move(board, X, direction, x, y, time):

    for i in range(X):
        if direction == 'R':
            dy = y + 1
            dx = x
        elif direction == 'L':
            dy = y - 1
            dx = x
        elif direction == 'U':
            dy = y
            dx = x - 1
        else:
            dy = y
            dx = x + 1

        if 0 == checkBoard(board, dx, dy):
            print(time+1)
            break
        else:
            time += 1
    return dx, dy, time

def snakeTurn(direction, C):
    if C == 'L':
        if direction == 'R':
            return 'U'
        elif direction == 'L':
            return 'D'
        elif direction == 'U':
            return 'L'
        else:
            return 'R'
    else:
        if direction == 'R':
            return 'D'
        elif direction == 'L':
            return 'U'
        elif direction == 'U':
            return 'R'
        else:
            return 'L'

for _ in range(L):
    X, C = input().strip().split()
    X = int(X)
    x, y, time = move(board, X, direction, x, y, time)
    direction = snakeTurn(direction, C)
    time += 1


# 뱀의 몸길이 업데이트 로직 수정, 뱀이 방향을 바꾸고 계속 움직이는 것도 time check 필요
```
