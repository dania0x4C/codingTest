# 최초 풀이

```python
import sys

pipe = '(((()(()()))(())()))(()())'

pipeCount = 0
pipeCut = 0

for i in range(len(pipe)):

    if pipe[i] == "(":                                                                                                                                                                                                                                                                                                        
        pipeCount += 1
    else:
        if pipe[i-1] == "(":
            pipeCount -= 1
            pipeCut += pipeCount
        else:
            pipeCount -= 1

print(pipeCut)
```

# 오답노트
1. 레이저인지 판단을 위해서 () 형태가 나왔는지 index로 확인을 했다.
2. 하나의 조건을 파악하지 못 했다는 것은 알았지만 정확한 인지를 못 함

# 정답

```python
import sys

pipe = sys.stdin.readline().strip()

pipeCount = 0
pipeCut = 0

for i in range(len(pipe)):

    if pipe[i] == "(":                                                                                                                                                                                                                                                                                                        
        pipeCount += 1
    else:
        pipeCount -= 1
        if pipe[i-1] == "(":
            pipeCut += pipeCount
        else:
            pipeCut += 1

print(pipeCut)
```
